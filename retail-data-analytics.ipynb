## Qns: 1. What is the trend in weekly sales across time - per year, per month


```python
# Sum of sales  across time 
sales_across_time = df_new.groupby(by = ['Date'], as_index = False)['Weekly_Sales'].sum()

plt.figure(figsize = (25,5))
plt.plot(sales_across_time.Date, sales_across_time.Weekly_Sales)
plt.title("Sum of Weekly sales across time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show

# Sales are generally constant, except on certain months (Nov, Dec)
```

```python
yearly_sales = df_new.groupby(df_new.Date.dt.year)['Weekly_Sales'].sum()/1000000
sns.barplot(x=yearly_sales.index,y=yearly_sales.values,data=df_new,palette='flare')
```

```python
yearly_sales = df_new.groupby(df_new.Date.dt.year)['Weekly_Sales'].sum()/1000000
sns.barplot(x = yearly_sales.index, y = yearly_sales.values, data = df_new, palette = 'rocket')
plt.title("Sum of Weekly Sales per year")
plt.ylabel("Sum of Weekly Sales (Millions)")
plt.xlabel("Years")

# Sum of Weekly sales is the highest in 2011 -- look into factors that may affect the sales
```

```python
# Sum of weekly sales each month, across the years 
monthly_sales = df_new.groupby(df_new.Date.dt.month)['Weekly_Sales'].sum()/1000000
sns.barplot(x = monthly_sales.index, y = monthly_sales.values, data = df_new, palette='flare')
plt.xlabel('Months')
plt.ylabel('Sum of Weekly Sales (Millions)')
plt.title("Sum of weekly sales per month")

# Sum of weekly sales is the highest in April, followed by October
```

```python
# Further deepdive into specific dates that drives highest weekly sales 
# Top 5 dates where weekly_sales are the highest 
sort_sales_across_time = sales_across_time.sort_values('Weekly_Sales', ascending = False) # sort in descending order 
sort_sales_across_time.head(5) 
```

# Qns 8: Trend of markdown across time 

#1.4: Graphical insights on markdown 

The graph indicates significant promotional activity in specific markdown categories, with MarkDown3 having a notable peak at the end of 2011.

Other markdowns, such as MarkDown1, MarkDown2, and MarkDown4, exhibit regular fluctuations, pointing to periodic promotional events.

MarkDown5 shows minimal activity, indicating it's likely a less frequently used markdown category.

```python
# Average Markdown across time 
df_markdown = df_new.groupby('Date').agg({'MarkDown1': 'mean', 'MarkDown2': 'mean', 'MarkDown3': 'mean', 'MarkDown4': 'mean', 'MarkDown5': 'mean'})
plt.figure(figsize = (30,10))

plt.plot(df_markdown.index, df_markdown.MarkDown1, label = 'MarkDown1')
plt.plot(df_markdown.index, df_markdown.MarkDown2, label = 'MarkDown2')
plt.plot(df_markdown.index, df_markdown.MarkDown3, label = 'MarkDown3')
plt.plot(df_markdown.index, df_markdown.MarkDown4, label = 'MarkDown4')
plt.plot(df_markdown.index, df_markdown.MarkDown5, label = 'MarkDown5')


plt.xlabel("Date")
plt.ylabel("Average Markdown ($)")
plt.title("Average Markdown across Time")
plt.legend(loc = 'best')
```

```python
# Markdown across months 
df_markdown_month = df_new.groupby(df_new.Date.dt.month).agg({'MarkDown1': 'mean', 'MarkDown2': 'mean', 'MarkDown3': 'mean', 'MarkDown4': 'mean', 'MarkDown5': 'mean'})
df_markdown_month.head()
df_markdown_month.plot(kind = 'bar', figsize = (20,10), rot = 0)
plt.xlabel("Months")
plt.ylabel('Average Markdown')
plt.title('Average Markdown per Month')

# Markdown3 is the highest in November 
```

```python
# Frequency of markdown 
df_markdown.hist(bins=10, figsize = (10,8), color = 'Purple')
plt.tight_layout()
plt.show()
```

The histograms indicate that while markdowns are commonly applied across different categories, most of them are relatively small in value.

Large markdowns are much less frequent but do occur occasionally, particularly in MarkDown1, MarkDown3, and MarkDown4.

The right-skewed nature of the distributions suggests that the retailer applies heavy discounts sparingly, likely during specific events or promotional periods.

Understanding the distribution of these markdowns can help in identifying patterns in pricing strategies and in planning future promotional activities.

# Correlation of factors and weekly sales 

```python

plt.figure(figsize = (16,6))
df_new = pd.DataFrame(df_new)
df_new = df.astype(float, errors = 'ignore')
heatmap = sns.heatmap(df_new.corr(), vmin = -1, vmax = 1, annot = True, cmap = 'BrBG')
heatmap.set_title('Correlation heatmap', fontdict = {'fontsize':18}, pad=12);

```

Markdown 1 and Markdown 4 can be positively correlated. It shows that they might be applied on the same time .

There is a negative correlation between Unemployment and CPI (-0.30). This suggests that higher unemployment rates are associated with lower CPI values, which could imply that economic downturns (with higher unemployment) lead to lower consumer spending and thus lower overall price levels.

There is a positive correlation between Weekly_Sales and Dept (0.15). This suggests that certain departments tend to have higher sales, possibly indicating that some departments perform consistently better across the stores.

The Temperature variable shows a slight negative correlation with IsHoliday (-0.16) and MarkDown3 (-0.20). This might imply that certain holidays or markdown events occur during colder periods.

There is a strong positive correlation between Fuel_Price and Date (0.75), which likely indicates that fuel prices have generally increased over time in the dataset.

Store has a slightly negative correlation with CPI (-0.21) and a positive correlation with Unemployment (0.21). This could indicate regional differences in economic conditions across different store locations.
